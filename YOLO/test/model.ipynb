```python
import cv2
from ultralytics import YOLO

# YOLOv8
model = YOLO("best.pt")  # 

# 
stream_url = "http://192.168.31.113:81/stream"  # URL
cap = cv2.VideoCapture(stream_url)

# 
if not cap.isOpened():
    print("Error: Could not open video stream")
    exit()

# 
while True:
    # 
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    # YOLOv8
    results = model(frame)

    # 
    annotated_frame = results[0].plot()

    # 
    cv2.imshow("YOLOv8 Stream Detection", annotated_frame)

    #  'q' 
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 
cap.release()
cv2.destroyAllWindows()
```
