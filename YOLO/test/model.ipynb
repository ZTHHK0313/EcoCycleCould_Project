```python
import cv2
from time import sleep, time
import pandas
from ultralytics import YOLO
import requests




# YOLOv8
model = YOLO("best.pt")  # 

# 
stream_url = "http://192.168.31.54:81/stream"  # URL
cap = cv2.VideoCapture(stream_url)

# 
if not cap.isOpened():
    print("Error: Could not open video stream")
    exit()

# 
while True:
    
    # 
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    # YOLOv8
    results = model(frame)

    # 
    annotated_frame = results[0].plot()

    # 
    cv2.imshow("YOLOv8 Stream Detection", annotated_frame)
    classes = [box.cls for box in results[0].boxes]  # 
    class_names = [model.names[int(cls)] for cls in classes]  # 
    
    if cv2.waitKey(1) & 0xFF == ord('a'):
        if len(class_names) > 0:
            res = max(set(class_names), key=class_names.count)
            print(res)
            try:
                if res == "Paper":
                    requests.get("http://192.168.31.161/api/1")
                elif res == "Plastic":
                    requests.get("http://192.168.31.161/api/2")
                elif res == "Metal":
                    requests.get("http://192.168.31.161/api/3")
            except:
                ...
            
            
        else:
            print("No detection")




    #  'q' 
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 
cap.release()
cv2.destroyAllWindows()
```
