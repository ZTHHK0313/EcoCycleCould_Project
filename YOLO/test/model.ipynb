```python
import cv2
from time import sleep, time
import pandas
from ultralytics import YOLO
import requests




# 加载训练好的YOLOv8模型
model = YOLO("best.pt")  # 替换为你的模型路径

# 设置视频流
stream_url = "http://192.168.31.54:81/stream"  # 替换为你的视频流URL
cap = cv2.VideoCapture(stream_url)

# 检查视频流是否成功打开
if not cap.isOpened():
    print("Error: Could not open video stream")
    exit()

# 实时检测循环
while True:
    
    # 读取视频帧
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    # 使用YOLOv8模型进行推理
    results = model(frame)

    # 在帧上绘制检测结果
    annotated_frame = results[0].plot()

    # 显示带注释的帧
    cv2.imshow("YOLOv8 Stream Detection", annotated_frame)
    classes = [box.cls for box in results[0].boxes]  # 获取所有检测框的类别索引
    class_names = [model.names[int(cls)] for cls in classes]  # 将索引转换为类别名称
    
    if cv2.waitKey(1) & 0xFF == ord('a'):
        if len(class_names) > 0:
            res = max(set(class_names), key=class_names.count)
            print(res)
            try:
                if res == "Paper":
                    requests.get("http://192.168.31.161/api/1")
                elif res == "Plastic":
                    requests.get("http://192.168.31.161/api/2")
                elif res == "Metal":
                    requests.get("http://192.168.31.161/api/3")
            except:
                ...
            
            
        else:
            print("No detection")




    # 按 'q' 键退出
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 释放资源
cap.release()
cv2.destroyAllWindows()
```
