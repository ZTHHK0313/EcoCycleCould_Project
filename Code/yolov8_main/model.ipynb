```python
import cv2
from time import sleep, time
import pandas
from ultralytics import YOLO
import requests




model = YOLO("best.pt")

stream_url = "http://192.168.1.249:81/stream"  
cap = cv2.VideoCapture(stream_url)

if not cap.isOpened():
    print("Error: Could not open video stream")
    exit()
ok = False

while True:
    
    
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

  
    results = model(frame)


    annotated_frame = results[0].plot()


    cv2.imshow("YOLOv8 Stream Detection", annotated_frame)
    classes = [box.cls for box in results[0].boxes]
    class_names = [model.names[int(cls)] for cls in classes]  
    
    
    
    if len(class_names) > 5:
        res = max(set(class_names), key=class_names.count)
        print(res)
        try:
            if res == "Paper":
                requests.get("http://192.168.1.156/api/1")
            elif res == "Plastic":
                requests.get("http://192.168.1.156/api/2")
            elif res == "Metal":
                requests.get("http://192.168.1.156/api/3")

        except:
            ...
            
            




    #  'q' 
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 
cap.release()
cv2.destroyAllWindows()
```
